```python
# some visualizations and testing on the keck tt data

import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from scipy import signal, optimize
```

```python
centroids = np.load('keck_tt/Centroid_n0088.npy')
centroids_1s = centroids[:1000]
```

```python
plt.figure(figsize=(10,10))
plt.plot(centroids_1s[:,0], centroids_1s[:,1], 'ko-')
lim = np.max(np.abs(centroids_1s))
plt.xlim((-lim,lim))
plt.ylim((-lim,lim))
```

```python
plt.figure(figsize=(10,10))
freqs, centroid_powers = signal.periodogram(centroids[:,0], fs=1000)
plt.loglog(freqs, centroid_powers)
```

```python
power_law_to_fit = np.vectorize(lambda f, a, b, c: a/(f + b)**2 + c)
par0 = [1, 1, 1]
par, cov = optimize.curve_fit(power_law_to_fit, freqs, centroid_powers, par0)
plt.semilogy(freqs, centroid_powers)
plt.semilogy(freqs, power_law_to_fit(freqs, *par))
```
