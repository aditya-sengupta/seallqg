```python
# generate pure turbulence with a known Cn2, and find the best parameters to control it

import numpy as np
from scipy import signal
from matplotlib import pyplot as plt

from aberrations import *

f_sampling = 1000

rms = lambda data: np.sqrt(np.mean(data ** 2))
psd = lambda data: signal.periodogram(data, fs=f_sampling)

heights = np.array([500, 1000, 2000, 4000, 8000, 16000])
velocities = np.array([10, 10, 10, 10, 10, 10])
Cn_squared = np.array([0.2283, 0.0883, 0.0666, 0.1458, 0.3350, 0.1350]) * 1e-12

mcmc_fit_steps = 2000

make_layers = lambda scale_cn2: make_custom_atmospheric_layers(heights, velocities, scale_cn2 * Cn_squared)(pupil_grid, L0=10)
make_tts = lambda scale_cn2: make_atm_data(steps=2000, layers=make_layers(scale_cn2))[:,0]
tts = make_tts(1)
```

```python
# by MCMC, find the best AR(3) model with which to control these TTs.
# Then change around the scale factor and see how that changes.

def metropolis_hastings(proposal_func, init_func, acceptance_prob,
                        num_iters, step=30):
    samples = []
    state = init_func()
    num_proposals = 0
    for i in range(num_iters):
        candidate = proposal_func(state)
        if np.random.uniform(0, 1) < min(1, acceptance_prob(state, candidate)):
            state = candidate
        if i % step == 0:
            samples.append(state)
    
    return samples

starting_state = lambda: np.array([0.6, 0.32, 0.08]) # the Keck standard integrator...sort of. 
# That ran on commands not centroids. But let's go with this.
def sample_candidate(sample):
    deltas = np.array([np.random.normal(0, 0.01), np.random.normal(0, 0.01), 0])
    deltas[2] = -sum(deltas)
    return sample + deltas

def spotsize(sample):
    return rms(np.convolve(sample, tts))

def score(sample, candidate):
    return spotsize(candidate) / spotsize(sample)

samples = metropolis_hastings(sample_candidate, starting_state, score, 10000, step=1)
```

```python
print(rms(np.convolve(starting_state(), tts)))
print(rms(np.convolve(samples[-1], tts)))
```
