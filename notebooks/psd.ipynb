## Aim

Try out the different periodogram methods on atm-vib data.

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import signal
from scipy.signal import periodogram, welch, lombscargle
from scipy.signal.windows import bartlett
%matplotlib inline
```

```python
ol = np.load("../data/sims/ol_atm_1_vib_3.npy")
```

```python
f, p = periodogram(ol[:,1], fs=1000)
plt.loglog(f, p)
```

```python
k = 5
plt.semilogy(f, np.convolve(np.hanning(2*k+1), p)[k:-k])
```

```python
k = 2
plt.semilogy(f, np.convolve(bartlett(2*k+1), p)[k:-k])
```

```python
f, p_w = welch(ol[:,1], fs=1000)
plt.semilogy(f, p_w)
```

```python
p_l = lombscargle(np.arange(0, 10, 0.001), ol[:,1], np.arange(0.5, 500, 0.5))
plt.loglog(np.arange(0.5, 500, 0.5), p_l)
```
